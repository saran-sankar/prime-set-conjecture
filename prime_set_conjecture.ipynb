```python
from itertools import chain, combinations

def powerset(iterable):
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

```python
N = 5
```

```python
E = {2}
```

```python
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return set(factors)
```

```python
%%time
for i in range(N):
    for subset in powerset(E):
        rho = 1
        for c in subset:
            rho = c*rho
        if rho>1:
            rho += 1
    
        for prime_factor in prime_factors(rho):
            E.add(prime_factor)
```

```python
P = []
def SieveOfEratosthenes(n): 
    prime = [True for i in range(n+1)] 
    p = 2
    while (p * p <= n): 
        if (prime[p] == True): 
            for i in range(p * p, n+1, p): 
                prime[i] = False
        p += 1
      
    for p in range(2, n): 
        if prime[p]: 
            P.append(p)
  
if __name__=='__main__': 
    n = 30
    SieveOfEratosthenes(100000)
```

```python
for i, prime in enumerate(P):
    if prime not in E:
        print("generated all of the first",i,"prime numbers. Next prime:", prime)
        break
```
